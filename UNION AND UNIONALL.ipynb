## UNION AND UNIONALL
PySpark union() and unionALL() transformation are used to merge two or more DF's of the same schema or structure 

```python
data1 = [
    (1, 'Alice', 'HR'),
    (2, 'Bob', 'IT'),
    (3, 'Charlie', 'HR'),
    (4, 'David', 'Finance'),
    (5, 'Eve', 'IT')
]
columns = ['id', 'name', 'dept']
df1 = spark.createDataFrame(data1, columns)
df1.show()
```

```python
data2 = [
    (3, 'Bob', 'IT'),
    (4, 'David', 'IT'),
    (5, 'Charlie', 'Finance')
 ]
columns = ['id', 'name', 'dept']
df2 = spark.createDataFrame(data2, columns)
df2.show()
```

```python
df1.union(df2).show()
```

```python
df1.unionAll(df2).show()
```
