## FILETER IN PYSPARK


```python
df = spark.read.csv("/Volumes/development/data/files/orders/orders.csv", header=True, inferSchema = True)
```

```python
display(df.filter(df.category == "Office Supplies"))
```

```python
display(df.filter((df.category == "Office Supplies" ) & (df.sales > 500)))
```

```python
display(df.filter((df.category == "Office Supplies") | (df.sales > 40)))
```

```python
display(df.filter(df.order_id.endswith(43)))
```

```python
display(df.filter(df.product_id.like("%AR%")))
```
