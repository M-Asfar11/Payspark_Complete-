**date_format in PySpark**

The date_format() function in PySpark is a powerful tool for transforming, formatting date columns, and converting dates to strings within a DataFrame.

This function allows you to convert date and timestamp columns into a specified string format, providing flexibility for various date manipulation tasks.

```python
from pyspark.sql import SparkSession, functions as f 
spark = SparkSession.builder.getOrCreate()

data = [
    (1, "2025-08-23"),
    (2, "2025-12-05"),
    (3, "2024-01-01"),
    (4, "2023-07-15")
]

df = spark.createDataFrame(data, ["id", "date"])
display(df)
```

```python
from pyspark.sql.functions import * 
df.select("id", "date", date_format("date", "yyyy/MM/dd")).show()
```

```python
from pyspark.sql.functions import * 
df.select("id", "date", date_format("date", "yyyy/MMMM/dd")).show()
```

```python
from pyspark.sql.functions import * 
df.select("id", "date", date_format("date", "dd/MM/yyyy")).show()
```
