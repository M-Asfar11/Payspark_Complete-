## What is groupBy() in PySpark

the groupBy() transformation groups rows that have the same values in specified columns. After grouping, you must apply aggregation functions (like, count, sum, avg, min, max).


```python
data = [
    (1, "Alice", "NY", 2000),
    (2, "Bob", "CA", 1500),
    (3, "Charlie", "NY", 3000),
    (4, "David", "CA", 2500),
    (5, "Eve", "TX", 1800),
    (6, "Frank", "TX", 2200),
]

columns = ["id", "name", "state", "salary"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
display(df.groupBy('state').count())
```

```python
display(df.groupBy('state').agg({'salary':'sum'}))
```

```python
display(df.groupBy('state').agg({'salary':'avg'}))
```

```python
display(df.groupBy('state').agg({'salary':'max'}))
```

```python
display(df.groupBy('state').agg({'salary':'min'}))
```
