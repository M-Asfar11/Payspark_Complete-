## StructType and StructField 
StructType -> Represents the entire schema of a DataFrame (like a container for all fields)

StructField -> Represents one single column/field inside that schema 

StructType is lika a list of StructField objects

StructField describes name, datatype, and nullability for one column

```python
from pyspark.sql.types import StructField, StructType, StringType, IntegerType
from pyspark.sql import SparkSession

schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])
data = [(1, "John", 25), (2, "Jane", 30), (3, None , 35)]
df = spark.createDataFrame(data, schema)

display(df)
df.printSchema()
```

```python
from pyspark.sql.types import StructField, StructType, StringType, IntegerType

schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("salary", IntegerType(), True)
])

df2 = spark.read.csv("employee.csv.", header= True, schema= schema)

df2.printSchema()
df2.show()
```
